```python
import csv
import pandas as pd
import seaborn as sb
```

**Se crea una variable con los valores asignados del archivo con pandas**

```python
df=pd.read_csv("data/Camas_UCI_diarias.csv")
```

**El comando shape se utiliza para mostrar la informacion sobre el tamaño del arreglo**

```python
df.shape
```

**La funcion describe muestra estadisticas para las columnas como por ejemplo la media, la mediana, el maximo, minimo, etc**

```python
df.describe()
```

**Se separan los valores de Camas UCI habilitadas del resto para poder graficar**

```python
cuh=df[df.Serie == 'Camas UCI habilitadas']
```

**Se dropean las variable sobrantes de la columna Serie**

```python
cuh.drop('Serie', axis=1, inplace=True)
```

**Se asigna la columna de Region como columna principal**

```python
cuh.set_index('Region',inplace=True)
```

**Se dropea la fila con el nombre Total para luego poder graficar**

```python
cuh.drop('Total', axis=0, inplace=True)
```

**Se procede a graficar los datos antes procesados de Camas UCI habilitadas (Se repite el mismo proceso para las demas secciones y graficos)**

```python
cuh.T.plot(figsize=(30,15), title='Camas UCI habilitadas', ylabel='Cantidad de camas', xlabel='Fechas')
sb.set_style('darkgrid')
sb.set_palette('pastel')
```

```python
cud=df[df.Serie == 'Camas UCI ocupadas COVID-19']
```

```python
cud.drop('Serie', axis=1, inplace=True)
```

```python
cud.set_index('Region',inplace=True)
```

```python
cud.drop('Total', axis=0, inplace=True)
```

```python
cud.T.plot(figsize=(30,15), title='Camas UCI ocupadas COVID-19', ylabel='Cantidad de camas', xlabel='Fechas')
sb.set_style('darkgrid')
sb.set_palette('pastel')
```
