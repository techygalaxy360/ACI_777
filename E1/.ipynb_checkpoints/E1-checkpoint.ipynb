```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

**Se procesan los datos en listas desde el csv para posteriormente graficarlos de forma mas sencilla**

```python
columnas=[]

with open('data/Camas_UCI_diarias.csv', encoding='utf-8') as archivo_csv:
    lector = csv.reader(archivo_csv, delimiter=',')
    header = next(lector)
    listaa = list(lector)
    result = []
    years = []
    for idx, col in enumerate(header[2:], start = 2):
        years.append(header[idx])
    for row in listaa:
        dates = []
        for idx, col in enumerate(row[2:], start = 2):
            dates.append(float(col))
        result.append(dict(region=row[0], section=row[1], years=years, dates=dates))
    
```

**Se utilizan los datos ya procesados para graficar los datos correspondientes a las camas UCI habilitadas**

```python
fig,ax = plt.subplots(figsize=(100,10))
for a in range(17):
    ax.set_xticklabels( result[a]['years'], rotation=90)
    plt.title(f"Seccion  {result[a]['section']}")
    ax.legend()
    plt.plot(result[a]['years'], result[a]['dates'],label=result[a]['region'])
```

**Se utilizan los datos ya procesados para graficar los datos correspondientes a las camas UCI ocupadas COVID-19**

```python
fig,ax = plt.subplots(figsize=(100,10))

    
for a in range(17,34):
    ax.set_xticklabels( result[a]['years'], rotation=90)
    plt.title(f"Seccion  {result[a]['section']}")
    ax.legend()
    plt.plot(result[a]['years'], result[a]['dates'],label=result[a]['region'])
```
